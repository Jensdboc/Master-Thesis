```python
import openai
client = openai.OpenAI(api_key = api_key)
import cv2
import base64
import os

video_path = "/project_ghent/Master-Thesis/video9.mp4"

gif_conf_root = "/project_ghent/Master-Thesis/gif/data/Confusion/"
gif_conf_paths = []
for gif_end in os.listdir(gif_conf_root)[:10]:
    gif_conf_paths.append(gif_conf_root + gif_end)
print(gif_conf_paths)

for gif_conf_path in gif_conf_paths:
    video = cv2.VideoCapture(gif_conf_path)
    base64Frames = []
    while video.isOpened():
        success, frame = video.read()
        if not success:
            break
        _, buffer = cv2.imencode(".jpg", frame)
        base64Frames.append(base64.b64encode(buffer).decode("utf-8"))
    
    video.release()
    
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "system",
                "content": """Does the response of the user indicate that the robots made a mistake, provoked confusion or any other elements that caused a disturbance in the conversation? 
                Describe the facial expressions and come up with a conclusion.
                Finally, answer with [CONFUSED] or [NEUTRAL]"""
              # "content": """Heeft de reactie van de gebruiker aangegeven dat de robots een fout hebben gemaakt, verwarring hebben veroorzaakt of andere elementen die verstoring in het gesprek hebben veroorzaakt? 
                # Antwoord alleen met '[JA]', '[NEE]' of '[MISSCHIEN]' in geval van onvoldoende informatie."""
            },
            {
                "role": "user", 
                "content": [{"image": frame} for frame in base64Frames[0:5]]
            }
        ]
    )
    print(response.choices[0].message.content)
    print("---------------")
```
